```python
import zipfile
file_name = zipfile.ZipFile('trash.zip', 'r')
file_name.extractall()
```

```python
# all images are load in variable x...
FILE='trash.zip'
img_name=[]
x_gray=[]
x_original = []
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
        #x_gray.append(cv2.imread(name, 0))
        x_gray.append(cv2.resize(cv2.imread(name, 0),(80,60)))
```

```python
y_names = []
for yy in img_name:
  name = yy[0]+yy[1]+yy[2]
  y_names.append(name)
```

```python
unique_list = []    
for yy in y_names:
  if yy not in unique_list:
    unique_list.append(yy)
```

```python
n = len(unique_list)
y = []
for yy in y_names:
  for i in range(n):
    if(yy == unique_list[i]):
      y.append(i)
```

```python
# view some images from the train_ds
num = len(y_names)
plt.figure(figsize=(16, 16))
for i in range(12):
  ax = plt.subplot(4, 4, i + 1)
  kk = random.randrange(num) 
  plt.imshow(x_gray[kk], cmap='gray')
  plt.title(y_names[kk])
  plt.axis("off")
```

```python
x_data = []
for xx in x_gray:
  x_data.append(xx.ravel())

df = pd.DataFrame(x_data)
print(df)
```

```python
nrows, ncols = df.shape
# view some images from the train_ds
num = nrows
plt.figure(figsize=(16, 16))
for i in range(12):
  ax = plt.subplot(4, 4, i + 1)
  kk = random.randrange(num)
  img_gray = np.array(df.iloc[kk,:]).reshape(60,80)
  plt.imshow(img_gray, cmap='gray')
  plt.title(y_names[kk])
  plt.axis("off")
```
